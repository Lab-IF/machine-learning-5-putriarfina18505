```python
import numpy as np

print("=== Tugas 1: Eksplorasi Array dan Matrix ===")
# 1. Array 20 angka random (0-100)
arr = np.random.randint(0, 101, 20)
print(f"Array: {arr}")

# 2. Statistik
print(f"Mean   : {np.mean(arr):.2f}")
print(f"Median : {np.median(arr):.2f}")
print(f"Std    : {np.std(arr):.2f}")
print(f"Min    : {np.min(arr)}")
print(f"Max    : {np.max(arr)}")

# 3. Matrix 5x5
print("\nMatrix 5x5 Random:")
matrix = np.random.randint(0, 101, (5, 5))
print(matrix)

# 4. Transpose
print("\nTranspose Matrix:")
print(matrix.T)

# 5. Determinant
print(f"\nDeterminant: {np.linalg.det(matrix):.2f}")
print("="*50)
```

```python
import pandas as pd

print("\n=== Tugas 2: Analisis Data dengan Pandas ===")
# Buat data 10 mahasiswa
data = {
    'NIM': [f'10{i}' for i in range(1, 11)],
    'Nama': ['Pina', 'Andi', 'Budi', 'Citra', 'Deni', 'Eka', 'Fani', 'Gito', 'Hana', 'Ian'],
    'Matematika': [85, 70, 90, 75, 80, 95, 65, 88, 72, 82],
    'Fisika': [80, 75, 85, 70, 78, 92, 60, 85, 75, 80],
    'Kimia': [90, 80, 88, 82, 85, 96, 70, 90, 78, 84]
}
df = pd.DataFrame(data)

# 1. Rata-rata per mata kuliah
print("Rata-rata Nilai per Mata Kuliah:")
print(df[['Matematika', 'Fisika', 'Kimia']].mean())

# 2. Rata-rata per mahasiswa
df['Rata_Rata'] = df[['Matematika', 'Fisika', 'Kimia']].mean(axis=1)

# 3. 5 Mahasiswa tertinggi
print("\n5 Mahasiswa dengan Nilai Tertinggi:")
print(df.nlargest(5, 'Rata_Rata').to_string(index=False))

# 4. Filter >= 80
print("\nMahasiswa dengan Rata-rata >= 80:")
print(df[df['Rata_Rata'] >= 80].to_string(index=False))

# 5. Export CSV
df.to_csv('Data_Mahasiswa_Pertemuan01.csv', index=False)
print("\nStatus: File CSV Berhasil Dibuat.")
print("="*50)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

print("\n=== Tugas 3: Visualisasi Data ===")
plt.figure(figsize=(12, 10))

# 1. Bar chart rata-rata MK
plt.subplot(2, 2, 1)
df[['Matematika', 'Fisika', 'Kimia']].mean().plot(kind='bar', color=['skyblue', 'salmon', 'lightgreen'])
plt.title('Rata-rata Nilai per Mata Kuliah')

# 2. Histogram Matematika
plt.subplot(2, 2, 2)
plt.hist(df['Matematika'], bins=5, color='purple', edgecolor='black')
plt.title('Distribusi Nilai Matematika')

# 3. Scatter plot MTK vs Fisika
plt.subplot(2, 2, 3)
plt.scatter(df['Matematika'], df['Fisika'], color='orange')
plt.xlabel('Matematika'); plt.ylabel('Fisika')
plt.title('Hubungan MTK dan Fisika')

# 4. Box plot semua nilai
plt.subplot(2, 2, 4)
sns.boxplot(data=df[['Matematika', 'Fisika', 'Kimia']])
plt.title('Box Plot Semua Nilai')

plt.tight_layout()
plt.show()
```

```python
from sklearn.datasets import load_wine

print("\n=== Tugas 4: Eksplorasi Dataset Wine ===")
# Load dataset
wine = load_wine(as_frame=True)
df_wine = wine.frame

# 1. Tampilkan 10 baris pertama
print("10 Baris Pertama Dataset:")
print(df_wine.head(10).to_string())

# 2. Statistik deskriptif
print("\nStatistik Deskriptif:")
print(df_wine.describe().to_string())

# 3. Visualisasi (2 viz)
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
df_wine['target'].value_counts().plot(kind='bar', color='coral')
plt.title('Distribusi Kelas Wine')

plt.subplot(1, 2, 2)
plt.scatter(df_wine['alcohol'], df_wine['color_intensity'], c=df_wine['target'], cmap='viridis')
plt.xlabel('Alcohol'); plt.ylabel('Color Intensity')
plt.title('Alcohol vs Color Intensity')
plt.show()

# 4. Interpretasi (Tulis manual di laporan atau print)
print("\nInterpretasi:")
print("Dataset Wine berisi hasil analisis kimia dari wine yang berasal dari tiga kultivar berbeda.")
print("Statistik deskriptif menunjukkan rentang nilai yang bervariasi pada tiap fitur kimia.")
print("Visualisasi menunjukkan adanya perbedaan konsentrasi alkohol dan intensitas warna pada tiap kelas target.")
```
